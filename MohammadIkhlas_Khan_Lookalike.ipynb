```python
import pandas as pd
```

```python
customers = pd.read_csv("C:/Users/hp/Downloads/Customers.csv")
products = pd.read_csv("C:/Users/hp/Downloads/Products.csv")
```

```python
customer_features = pd.get_dummies(customers, columns=['Region'], prefix='Region')
customer_features = customer_features.drop(['CustomerName', 'SignupDate'], axis=1)
```

```python
products_encoded = pd.get_dummies(products, columns=['Category'], prefix='Category')
for col in products_encoded.columns[1:]: 
    customer_features[col] = 0
```

```python
customer_features
```

```python
customer_features.set_index('CustomerID', inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(customer_features)
similarity_df = pd.DataFrame(similarity_matrix, index=customer_features.index, columns=customer_features.index)
```

```python
def get_top_3_similar(customer_id, similarity_df):
    if customer_id not in similarity_df.index:
        return []
    similar_customers = similarity_df[customer_id].sort_values(ascending=False).iloc[1:4]  # Skip self-similarity
    return [(cust_id, score) for cust_id, score in similar_customers.items()]

lookalike_results = {}
for customer_id in customers['CustomerID'][:20]:
    lookalike_results[customer_id] = get_top_3_similar(customer_id, similarity_df)

lookalike_df = pd.DataFrame({
    'CustomerID': lookalike_results.keys(),
    'Lookalikes': lookalike_results.values()
})

print(lookalike_df.head())
```

```python
lookalike_df.to_csv("Lookalike.csv", index=False)
```
