```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cust=pd.read_csv("C:/Users/hp/Downloads/Customers.csv")
prod=pd.read_csv("C:/Users/hp/Downloads/Products.csv")
trans=pd.read_csv("C:/Users/hp/Downloads/Transactions.csv")
```

```python
cust
```

```python
prod
```

```python
trans
```

```python
cust_trans=pd.merge(cust,trans,on='CustomerID',how='inner')
```

```python
cust_trans
```

```python
full_data=pd.merge(cust_trans,prod,on='ProductID',how='inner')
```

```python
full_data
```

```python
full_data.drop('Price_y',axis=1,inplace=True)
```

```python
full_data
```

```python
top_products=full_data.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(10)
```

```python
top_products
```

```python
top_products.plot(kind='bar')
plt.title('Best Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
revenue_by_region=full_data.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)
```

```python
revenue_by_region
```

```python
revenue_by_region.plot(kind='pie',autopct='%1.1f%%',colors=['red','orange','yellow','green'])
plt.title('Revenue Distribution by Region')
plt.ylabel('')
plt.show()
```

```python
revenue_by_category=full_data.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)
```

```python
revenue_by_category
```

```python
revenue_by_category.plot(kind='bar')
plt.title('Revenue By Product Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
popular_products_by_region=full_data.groupby(['Region','ProductName'])['Quantity'].sum().reset_index()
```

```python
popular_products=popular_products_by_region.sort_values(['Region','Quantity'],ascending=[True,False])
```

```python
topproducts_by_region=popular_products.groupby('Region').head(1)
```

```python
topproducts_by_region
```

```python
plt.bar(topproducts_by_region['Region'], topproducts_by_region['Quantity'])

for i, product in enumerate(topproducts_by_region['ProductName']):
    plt.text(i, topproducts_by_region['Quantity'].iloc[i] + 10, product,ha='left',rotation=45,fontsize=10)

plt.title('Most Popular Product in Each Region')
plt.xlabel('Region')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_transaction_per_customer = full_data.groupby('CustomerID')['TotalValue'].mean().reset_index()
avg_transaction_per_customer.columns = ['CustomerID', 'AvgTransactionValue']
avg_transaction_per_customer.head()
```

```python
plt.hist(avg_transaction_per_customer['AvgTransactionValue'], bins=20, color='purple')
plt.title('Distribution of Average Transaction Value per Customer')
plt.xlabel('Average Transaction Value (USD)')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```
